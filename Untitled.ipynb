```python
#import data from csv
df = pd.read_csv('base.csv' , header=1, index_col=0)
#import from shoot.csv
df_shoot = pd.read_csv('shooting.csv' , header=1, index_col=0)
#import from pass.csv
df_pass = pd.read_csv('passing.csv' , header=1, index_col=0)
#import from goal_creation.csv
df_goal_creation = pd.read_csv('goal_creation.csv' , header=1, index_col=0)
#import from defence.csv
df_defence = pd.read_csv('defensive.csv' , header=1, index_col=0)
#import from possession.csv
df_possession = pd.read_csv('possession.csv' , header=1, index_col=0)

#merge all into df dataframe using name as index
# Merge all DataFrames on 'Player'
df = pd.merge(df, df_shoot, on='Player', suffixes=('', '_shoot'))
df = pd.merge(df, df_pass, on='Player', suffixes=('', '_pass'))
df = pd.merge(df, df_goal_creation, on='Player', suffixes=('', '_goal_creation'))
df = pd.merge(df, df_defence, on='Player', suffixes=('', '_defence'))
df = pd.merge(df, df_possession, on='Player', suffixes=('', '_possession'))
# Remove duplicate columns by keeping only the original ones
columns_to_keep = []
for column in df.columns:
    if column.endswith(('_shoot', '_pass', '_goal_creation', '_defence', '_possession')):
        original_column = column.rsplit('_', 1)[0]
        if original_column not in df.columns:
            df.rename(columns={column: original_column}, inplace=True)
    else:
        columns_to_keep.append(column)

df = df[columns_to_keep]
df['Age'] = df['Age'].str[:2]
#create a new column with the age as an int
df['Age'] = df['Age'].astype(int)
#make nation only the last 3  letters of the nation column
df['Nation'] = df['Nation'].str[-3:]
#make the nation column a string
df['Nation'] = df['Nation'].astype(str)


```

```python
def ascii_table(data):
    # Split data into rows
    rows = data.strip().split('\n')

    # Extract header and data
    header = rows[0].split()
    data = [row.split() for row in rows[1:]]

    # Determine column widths
    col_widths = [max(len(row[i]) for row in data) for i in range(len(header))]

    # Build table
    table = ''
    table += '+'.join(['-'*(w+2) for w in col_widths]) + '+\n'
    table += '| ' + ' | '.join(header[i].ljust(col_widths[i]) for i in range(len(header))) + ' |\n'
    table += '+'.join(['-'*(w+2) for w in col_widths]) + '+\n'
    for row in data:
        table += '| ' + ' | '.join(row[i].ljust(col_widths[i]) for i in range(len(header))) + ' |\n'
    table += '+'.join(['-'*(w+2) for w in col_widths]) + '+'

    return table
```
